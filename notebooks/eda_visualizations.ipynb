```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
markt_analytics = pd.read_csv("../data/processed/marketing_campaign_cleaned.csv")
```

```python
# Spending Behavior by Marital Status
plt.figure(figsize=(10,6))
sns.barplot(data=markt_analytics, x='Marital_Status', y='Total_Spent', estimator='mean')
plt.title('Average Total Spending by Marital Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap (Top Spending Factors)
plt.figure(figsize=(12,8))
sns.heatmap(
    markt_analytics.corr(numeric_only=True)[['Total_Spent']].sort_values(by='Total_Spent', ascending=False),
    annot=True, cmap='coolwarm'
)
plt.title('Correlation with Total Spending')
plt.xlabel('Total Spending (€)')
plt.ylabel('Feature')
plt.show()
```

Heatmap shows how each numerical feature is correlated with Total_Spent. The values range from:

    🔴 +1.0 (Strong Positive Correlation): feature increases as spending increases

    🔵 -1.0 (Strong Negative Correlation): decreases as spending increases

MntWines	+0.89	Higher wine spending → higher total spend

Kidhome	-0.56	More kids at home → lower total spend

Total_Spent	1.00	(Itself — always 1.0)

```python
# Distribution of Income
plt.figure(figsize=(8,6))
sns.histplot(markt_analytics['Income'], kde=True, bins=30)
plt.title('Customer Income Distribution')
plt.xlabel('Income (€)')
plt.ylabel('Number of Customers')
plt.show()
```
